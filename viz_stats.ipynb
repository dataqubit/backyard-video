```python
import pandas as pd
import altair as alt
```

```python
import seaborn as sns
```

```python
alt.data_transformers.enable("vegafusion")
```

```python
import matplotlib.pyplot as plt
```

```python
def render_motion_stats_graph(stats_csv_path):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(stats_csv_path)

    # Create a time-series chart using Altair
    chart = alt.Chart(df).mark_line(point=True).encode(
        x=alt.X('timestamp:T', title='Timestamp'),
        y=alt.Y('rectangle_size:Q', title='Motion Detection Rectangle Size'),
        tooltip=['timestamp:T', 'rectangle_size:Q']
    ).properties(
        title='Time-Series Motion Detection Stats',
        width=800,
        height=400
    ).configure_title(
        fontSize=20,
        anchor='start',
        color='black'
    ).configure_axis(
        labelFontSize=12,
        titleFontSize=14
    ).configure_point(
        size=60,
        fillOpacity=0.6
    )

    return chart
```

```python
def seaborn_render_motion_stats_plot(stats_csv_path):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(stats_csv_path)

    # Convert the timestamp to datetime
    df['timestamp'] = pd.to_datetime(df['timestamp'])

    # Set the style of the plot
    sns.set(style="darkgrid")

    # Create the plot
    plt.figure(figsize=(14, 7))
    plot = sns.lineplot(x='timestamp', y='rectangle_size', data=df, marker='o')

    # Set plot title and labels
    plot.set_title('Time-Series Motion Detection Stats', fontsize=20)
    plot.set_xlabel('Timestamp', fontsize=14)
    plot.set_ylabel('Motion Detection Rectangle Size', fontsize=14)

    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45)

    # Show the plot
    plt.tight_layout()
    plt.show()
```
